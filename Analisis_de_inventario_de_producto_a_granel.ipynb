```python
#Identificar los rotulos que no estan en ambas bases de datos, es decir no tienen una pareja para ser comparados
rotulos_no_repetidos = []

for index, row in df.iterrows():
    id_rotulo_df1 = row['ID Rotulo']
    if id_rotulo_df1 not in df2['ID Rotulo'].values:
        rotulos_no_repetidos.append((id_rotulo_df1, "df1"))

for index, row in df2.iterrows():
    id_rotulo_df2 = row['ID Rotulo']
    if id_rotulo_df2 not in df['ID Rotulo'].values:
        rotulos_no_repetidos.append((id_rotulo_df2, "df2"))

rotulos_no_repetidos
```

```python
#comparar ambos data frames
merged_df = pd.merge(df, df2, on='ID Rotulo', suffixes=('_df1', '_df2'))

comparison_columns = ['GramosGranel', 'JuegosGranel']
for col in comparison_columns:
    merged_df[f'{col}_Difference'] = merged_df[f'{col}_df1'] - merged_df[f'{col}_df2']

merged_df[['ID Rotulo', 'GramosGranel_df1', 'GramosGranel_df2', 'GramosGranel_Difference', 'JuegosGranel_df1', 'JuegosGranel_df2', 'JuegosGranel_Difference']]
```

```python
#comparar ambos data frames
merged_df = pd.merge(df, df2, on='ID Rotulo', suffixes=('_df1', '_df2'))

comparison_columns = ['GramosGranel', 'JuegosGranel']
for col in comparison_columns:
    merged_df[f'{col}_Difference'] = abs(merged_df[f'{col}_df1'] - merged_df[f'{col}_df2'])



merged_df[['ID Rotulo', 'GramosGranel_df1', 'GramosGranel_df2', 'GramosGranel_Difference', 'JuegosGranel_df1', 'JuegosGranel_df2', 'JuegosGranel_Difference']]
```

```python
lista_de_0 = [38403.0, 43263.0, 44134.0, 43653.0, 22188.0, 27727.0, 43765.0, 42626.0, 44942.0, 38650.0, 35276.0, 38828.0]

merged_df.drop(merged_df[merged_df['ID Rotulo'].isin(lista_de_0)].index, inplace=True)
```

```python
# calcular el porcentaje de variacion 'GramosGranel' y 'JuegosGranel'
merged_df['GramosGranel_PercentageVariation'] = round(abs(100 - (merged_df['GramosGranel_df1'] / merged_df['GramosGranel_df2']) * 100),3)
merged_df['JuegosGranel_PercentageVariation'] = round(abs(100 - (merged_df['JuegosGranel_df1'] / merged_df['JuegosGranel_df2']) * 100),3)


merged_df[['ID Rotulo', 'GramosGranel_df1', 'GramosGranel_df2', 'GramosGranel_Difference', 'GramosGranel_PercentageVariation',
                 'JuegosGranel_df1', 'JuegosGranel_df2', 'JuegosGranel_Difference', 'JuegosGranel_PercentageVariation']]
#con esta comparacion nos damos cuenta hay diferencias entre ambos set de datos,minetras el percentageVaration sea mas cercano a 100 los datos son mas diferentes
```

```python
sj = merged_df[merged_df['GramosGranel_PercentageVariation'] > 2000]
sj
```

```python
# Gráfico de dispersión
plt.figure(figsize=(10, 6))

# Datos de df1
plt.scatter(df['GramosGranel'], df['JuegosGranel'], color='blue', label='df1', alpha=0.6)

# Datos de df2
plt.scatter(df2['GramosGranel'], df2['JuegosGranel'], color='red', label='df2', alpha=0.3)

# Etiquetas y título
plt.xlabel('GramosGranel')
plt.ylabel('JuegosGranel')
plt.title('Relación entre GramosGranel y JuegosGranel (df1 vs df2)')
plt.legend()

plt.show()
```

```python
# Seleccionamos un subconjunto de filas para no saturar el gráfico

subset_df = df[df['ID Rotulo'].isin([13511.0,16980.0,32601.0,41782.0,13299.0])] # Primeras 5 filas de df1
subset_df2 = df2[df2['ID Rotulo'].isin([13511.0,16980.0,32601.0,41782.0,13299.0])]  # Primeras 5 filas de df2

# Configuramos el gráfico
fig, ax = plt.subplots(figsize=(12, 6))

# Posiciones de las barras
indices = range(len(subset_df))

# Barras para GramosGranel
ax.bar(indices, subset_df['GramosGranel'], width=0.4, label='GramosGranel (df1)', align='center')
ax.bar([i + 0.4 for i in indices], subset_df2['GramosGranel'], width=0.4, label='GramosGranel (df2)', align='center')



# Etiquetas y título
ax.set_xticks([i + 0.2 for i in indices])
ax.set_xticklabels(subset_df['ID Rotulo'], rotation=90)
ax.set_xlabel('ID')
ax.set_ylabel('Valores')
ax.set_title('Comparación de GramosGranel y JuegosGranel entre df1 y df2')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

# Barras para JuegosGranel
ax.bar(indices, subset_df['JuegosGranel'], width=0.4, label='JuegosGranel (df1)', align='edge',color="red")
ax.bar([i + 0.4 for i in indices], subset_df2['JuegosGranel'], width=0.4, label='JuegosGranel (df2)', align='edge',color="green")

# Etiquetas y título
ax.set_xticks([i + 0.2 for i in indices])
ax.set_xticklabels(subset_df['ID Rotulo'], rotation=90)
ax.set_xlabel('ID')
ax.set_ylabel('Valores')
ax.set_title('Comparación de GramosGranel y JuegosGranel entre df1 y df2')
ax.legend()

plt.tight_layout()
plt.show()
```
